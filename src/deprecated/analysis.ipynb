```python
from deeppower import DeepPower
from preprocess import Preprocess
```

```python
import time

power_estimation = DeepPower(
        data_file="../data/20200813-2012-merged.csv",
        hist_size=100,
        train_split=0.6,
        reverse_train_split=False,
        net="cnn",
        n_epochs=10,
        verbose=False,
        time_id=time.strftime("%Y%m%d-%H%M%S")
)
```

```python
power_estimation.preprocess(features=[], remove_features=["heartrate", "abdomen"])
power_estimation.build_model()
```

```python
power_estimation.fit(verbose=0)
```

```python
power_estimation.predict()
power_estimation.plot_prediction_plotly()
```
